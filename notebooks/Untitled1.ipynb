```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from scipy import linalg as la
```

```python
# Problem 2 (standing orbits)

# load in the data and assign to understandable variables
orb = np.load('orbits.npz',)
merc_orb = orb['mercury']
ven_orb = orb['venus']
earth_orb = orb['earth']
mars_orb = orb['mars']

# set the figure for 3d Plots
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim((-1.3,1.3))
ax.set_ylim((-1.3,1.3))
ax.set_zlim((-1.3,1.3))
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")


# plot the sun at the origin
ax.scatter3D(0,0,0, color='yellow', marker='o', lw=8, label='Sun')

# plot the standing location of mercury and its orbit
ax.plot3D([merc_orb[0,0]],[merc_orb[0,1]],[merc_orb[0,2]],color='gray',marker='.',lw=2)
ax.plot3D(merc_orb[:,0],merc_orb[:,1],merc_orb[:,2], 'gray', lw=.75, label='Mercury')

# plot the standing location of Venus and its orbit
ax.plot3D([ven_orb[0,0]],[ven_orb[0,1]],[ven_orb[0,2]],color='orange',marker='.',lw=2)
ax.plot3D(ven_orb[:,0],ven_orb[:,1],ven_orb[:,2], 'orange', lw=.75, label='Venus')

# plot the standing location of the Earth and its orbit
ax.plot3D([earth_orb[0,0]],[earth_orb[0,1]],[earth_orb[0,2]],color='blue',marker='.',lw=2)
ax.plot3D(earth_orb[:,0],earth_orb[:,1],earth_orb[:,2], 'b', lw=.75, label='Earth')

# plot the standing location of Mars and its orbit
ax.plot3D([mars_orb[0,0]],[mars_orb[0,1]],[mars_orb[0,2]],color='red',marker='.',lw=2)
ax.plot3D(mars_orb[:,0],mars_orb[:,1],mars_orb[:,2], 'r', lw=.75, label='Mars')


ax.legend(loc='upper right', prop={'size': 5})
ax.set_title("Orbits of the inner Planets")
plt.show()
```
